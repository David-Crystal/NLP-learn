```python
def get_prob(word):
    esp = 1/frequences_sum #未出现的词视为出现过1次
    if word in words_count:
        return words_count[word] / frequences_sum
    else:
        return esp
```

将一句话出现的概率用多个词汇概率乘积来近似表征

## 1-Gram 一个单词的自旋~独立事件 无其他联系 

$$ Pr(S) = Pr(w_1\cdot w_2\cdots w_n)\sim Pr(w_1)\cdot Pr(w_2)\cdots Pr(w_n) $$

我今天起晚与太平洋对岸大的女孩今早去喝星巴克之间的联系

```python
def language_model_one_gram(string):
    words = cut(string)
    return product(get_prob(w) for w in words)
```

```python
language_model_one_gram('荣格好牛逼')
```

```python
sentences = """
自性整合
暗夜海洋之旅
赫尔墨斯
伟大的灵魂都是雌雄同体
""".split()
```

```python
for s in sentences:
    print(s, language_model_one_gram(s))
```

```python
need_compared = [
    "今天晚上我请你吃大餐,我们一起吃战斧牛排 明天晚上你请我吃大餐,我们一起吃澳洲龙虾",
    "真是一只可爱的小鸡 真是一只可爱的小猫",
    "我去吃火锅,今晚 今晚我去吃火锅",
]
for s in need_compared:
    s1, s2 = s.split()
    p1, p2 = language_model_one_gram(s1), language_model_one_gram(s2)
    
    better = s1 if p1>p2 else s2
    
    print('{} is more possible'.format(better))
    print('-'*4 + '{} with probility {}'.format(s1, p1))
    print('-'*4 + '{} with probility {}'.format(s2, p2))
```

## 2-Gram

### Get the combination probability of 2-grams

$$ Pr(S) = Pr(w_1\cdot w_2\cdots w_n)\sim Pr(w_1)\cdot Pr(w_2|w_1)\cdots Pr(w_n|w_{n-1}) $$

$$ Pr(w_2|w_1) = \frac{Pr(w_1\cdot w_2)}{Pr(w_1)} $$

```python
valid_tokens = [str(t) for t in valid_tokens]
```

```python
all_2_grams_words = [''.join(valid_tokens[i:i+2]) for i in range(len(valid_tokens[:-2]))]
```

```python
all_2_grams_words
```

```python
_2_gram_sum = len(all_2_grams_words)
_2_gram_counter = Counter(all_2_grams_words)

def get_combination_prob(w1, w2):
    if w1 + w2 in _2_gram_counter: return _2_gram_counter[w1+w2] / _2_gram_sum
    else:
       return 1 / _2_gram_sum
```

```python
get_combination_prob('自','本周')
```

```python
get_combination_prob('去','北京')
```

```python
def get_prob_2_gram(w1, w2):
    return get_combination_prob(w1, w2) / get_prob(w1)
```

```python
get_prob_2_gram('去','无锡')
```

```python
get_prob_2_gram('去','北京')
```

```python
def language_model_of_2_gram(sentence):
    sentence_probability = 1

    words = cut(sentence)
    
    for i, word in enumerate(words):
        if i==0:
            prob = get_prob(word)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous, word)
        #ic(prob)
        sentence_probability *= prob
    
    return sentence_probability
```

## Sentence generation

```python
grammar = """
sentence => noun_phrase verb_phrase 
noun_phrase => Article Adj* noun belong
belong => null | de property
de => 的
property => 眼睛 | 裙子 | 胳膊 | 尾巴
Adj* => null | Adj Adj*
verb_phrase => verb noun_phrase
Article =>  一个 | 这个
noun =>   女人 |  篮球 | 桌子 | 小猫
verb => 看着   |  坐在 |  听着 | 看见
Adj =>   蓝色的 |  好看的 | 小小的
"""
```

```python
import random
```

```python
def parse_grammar(grammar_str, sep='=>'):
    grammar = {}
    for line in grammar_str.split('\n'): 
        line = line.strip()
        if not line: continue
        
        target, rules = line.split(sep)
        
        grammar[target.strip()] = [r.split() for r in rules.split('|')]
    
    return grammar
```

```python
def gene(grammar_parsed, target='sentence'):
    if target not in grammar_parsed: return target
    
    rule = random.choice(grammar_parsed[target])
    return ''.join(gene(grammar_parsed, target=r) for r in rule if r!='null')
```

```python
g = parse_grammar(grammar)
```

```python
random_generated = [gene(g) for _ in range(100)]
```

```python
sorted(random_generated, key=language_model_of_2_gram, reverse = True)
```

## Get 3-Gram

### Get the combination probability of 3-grams

$$ Pr(S) = Pr(w_1\cdot w_2\cdots w_n)\sim Pr(w_1)\cdot Pr(w_2|w_1, w_3)\cdots Pr(w_n|w_{n-1}) $$

$$ Pr(w_2|w_1, w_3) = \frac{Pr(w_1\cdot w_2\cdot w_3)}{Pr(w_1, w_3)} $$

```python
valid_tokens = [str(t) for t in valid_tokens]
```

```python
valid_tokens
```

```python
all_3_grams_words = [''.join(valid_tokens[i:i+3]) for i in range(len(valid_tokens[:-3]))]
```

```python
all_3_grams_words
```

```python
_3_gram_sum = len(all_3_grams_words)
_3_gram_counter = Counter(all_3_grams_words)

def get_combination_3_prob(w1, w2, w3):
    if w1 + w2 + w3 in _3_gram_counter: return _3_gram_counter[w1+w2+w3] / _3_gram_sum
    else:
       return 1 / _3_gram_sum
```

```python
get_combination_3_prob('你', '去', '吃饭')
```

```python
get_combination_3_prob('本周', '6', '月')
```

```python
def get_prob_3_gram(w1, w2, w3):
    return get_combination_3_prob(w1, w2, w3) / get_combination_prob(w1, w3)
```

```python
def language_model_of_3_gram(sentence):
    sentence_probability = 1

    words = cut(sentence)
    
    for i, word in enumerate(words[:-1]):
        if i==0:
            prob = get_prob(word)
        else:
            previous = words[i-1]
            after = words[i+1]
            prob = get_prob_3_gram(previous, word, after)
       #ic(word)    
       #ic(prob)
        sentence_probability *= prob
    previous=words[i]
    sentence_probability *= get_prob_2_gram(previous, words[i+1])
    
    return sentence_probability
```
